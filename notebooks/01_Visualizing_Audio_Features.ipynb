```python
sample_gunshot_path = '/content/drive/MyDrive/UrbanSound8k/sample_of_urbansound/fold1/*.wav'
sample_gunshot_path = glob.glob(sample_gunshot_path)
#print(type(sample_gunshot_path))
#print(sample_gunshot_path[0])
sample_gunshot_path = sample_gunshot_path[0]
```

```python
signal, sr = librosa.load(sample_gunshot_path,sr=22050)
plt.figure(figsize = (9,9))
librosa.display.waveplot(signal,sr=sr)

plt.style.use('ggplot')
plt.ylabel('Amplitude')
plt.xlabel('Time(s)')
#plt.plot(signal, sr)
plt.title('Raw Sound(Gunshot)', fontsize=16)
plt.show()
```

#**Feature**'s Plot

```python
mel_feature = librosa.feature.melspectrogram(y=signal, sr =sr)
mfcc_feature = librosa.feature.mfcc(y=signal,sr=sr)
chroma_stft_feature = librosa.feature.chroma_stft(y=signal,sr=sr)
#tonnetz = librosa.feature.tonnetz(y=signal,sr=sr)
```

#mel

```python
fig,ax = plt.subplots(figsize=(7,5))
ax.set_title('Mel spectrogram(gunshot)')
librosa.display.specshow(mel_feature,y_axis='mel', x_axis='time')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
#print(mel_feature)
```

```python
fig,ax = plt.subplots(figsize=(7,5))
ax.set_title('MFCC spectrogram(gunshot)')
librosa.display.specshow(mfcc_feature,y_axis='mel', x_axis='time')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```

```python
fig,ax = plt.subplots(figsize=(7,5))
ax.set_title('MFCC spectrogram(gunshot)')
librosa.display.specshow(chroma_stft_feature,y_axis='mel', x_axis='time')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```
